# Code for scarping the Data

```python
# Scrapping the Web Page

titles = []
ratings = []
prices = []
features = []
offers=[]
exchanges=[]



for i in range(1, 20):
    URL ='https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=b4a18f5a-186b-40c5-a62e-9a6c9425b420&page{}'. format(i) 
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for container in soup.find_all('div', attrs={'class' : '_2kHMtA'}):

        product = container.find('div', attrs={'class' : '_4rR01T'})
        if product is None:
            titles.append(np.NaN)
        else:
            titles.append(product.text)

        price = container.find('div', attrs={'class' : '_30jeq3 _1_WHN1'})
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)

        rating = container.find('div', attrs={'class' : '_3LWZlK'})
        if rating is None:
            ratings.append(np.NaN)
        else:
            ratings.append(rating.text)

        feature = container.find('ul', attrs={'class' : '_1xgFaf'})
        if feature is None:
            features.append(np.NaN)
        else:
            features.append(feature.text)
            
            
        offer = container.find('div', attrs={'class' : '_3Ay6Sb'})
        if offer is None:
            offers.append(np.NaN)
        else:
            offers.append(offer.text)
            
         
        exchange = container.find('div', attrs={'class' : '_3xFhiH'})
        if exchange is None:
            exchanges.append(np.NaN)
        else:
            exchanges.append(exchange.text)
            
        
                     
```

```python
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(features))
print(len(offers))
print(len(exchanges))
```

# Outlier check

 #Mean+3SD rule
 #Percentile rule
 #1.5*IQR rule

```python
m=df.MRP.mean()
sd=np.std(df.MRP)
out=df.loc[df.MRP>m+3*sd,"MRP"].count()

print(f"Mean+3SD  : {m+3*sd}")
print(f"Outliers  : {out}")
print(f"Outliers% : {np.round((out/df.MRP[0])*100,2)}%")
```

```python
p99=np.percentile(df.MRP,99)
out=df.loc[df.MRP>p99,"MRP"].count()

print(f"P99       : {p99}")
print(f"Outliers  : {out}")
print(f"Outliers% : {np.round((out/df.MRP[0])*100,2)}%")
```

```python
sns.boxplot(np.where(df.MRP>p99,p99,df.MRP))
```

```python
p98=np.percentile(df.MRP,98)
out=df.loc[df.MRP>p98,"MRP"].count()

print(f"P98       : {p98}")
print(f"Outliers  : {out}")
print(f"Outliers% : {np.round((out/df.shape[0])*100,2)}%")
```

```python
sns.distplot(np.where(df.MRP>p98,p98,df.MRP))
plt.title("98 Percentile")
plt.show()
```
